### **Inventory Management System**

```python
product_inventory = {} #A dictionary that stores the products' name and quantity
```

```python
#Function that add a new product to the inventory record
def add_products():
    try:
        while True:
            product_name = input("Enter the product name") #promt the user to input the product name
            if product_name in product_inventory: #Check if the product already exist in the product_inventory record
                print(f"{product_name} details already exist")
            else: #If the product doesn't exist before
                product_quantity = int(input("What's the quantity of the product?")) #prompt the user to enter the quantity of product
                if product_quantity > 0: #To make sure the user doen't input a negative number 
                    product_inventory[product_name] = product_quantity #add the product name as the key alongside with the quantity as the value in the product inventory dictionary
                    print(f"{product_name} details added successfully!")
                else:
                    print("Quantity should be positive") #Tells the user that the product quantity should be positive if negative quantity is entered
            opr = input("WIll you like to add any other product?")
            if opr == "No":
                break
    except ValueError:
        print("Invalid Quantity")
```

```python
#A function to search for any product record in the inventory record
def view_product():
    product_name = input("Enter the product name") #promt the user to input the product name
    if product_name in product_inventory: #Check if the product has been recorded in the product inventory
        product_quantity = product_inventory[product_name] #assigning the value to key(product)
        print(f"Product Name: {product_name}") #print the product name
        print(f"Quantity: {product_quantity}") #print the product quantity
    else:
        print("This product details has not been added")
```

```python
#A function to view all the products' records
def view_inventory():
    if len(product_inventory) ==0: ## To check first if the dictionary created is empty or not
        print("No any record has been added")
    else:
        for product in product_inventory: #A loop through al the products
            print(f"Product Name: {product}")
            product_quantity = product_inventory[product]
            print(f"Quantity: {product_quantity}\n")
```

```python
# Function to update a product quantity
def update_product_record():
    try:
        product_name = input("Enter the product name")
        if product_name in product_inventory:
            new_product_quantity = int(input("What's the new quantity?"))
            if new_product_quantity > 0:
                product_inventory[product_name] = new_product_quantity
                print("Updated successfully!")
            else:
                print("Quantity should be positive")
        else:
            print("This product details has not been added")
    except ValueError:
        print("Invalid input")
        
```

```python
#Function to remove a product record
def remove_product_record():
    product_name = input("Which product will you like to remove?") #asks for the product name
    if product_name in product_inventory: #Check if the product is prsent in the dictionary (Inventory record)
        del product_inventory[product_name] #Deleting the product record from the dictionary
        print("Product removed successfully!")
    else: 
        print("The product has not been added before")
```

```python
#Function for the inventory statistics
def inventory_statistics():
    if len(product_inventory) ==0: # To check first if the dictionary created is empty or not
        print("No any record has been added")
    else:
        total_number_of_products = len(product_inventory.keys())
        highest_quantity = max(product_inventory.values()) #assigning to the highest quantity the maximum of the dictionary values (which is the quantity of products)
        lowest_quantity = min(product_inventory.values()) #assigning to the lowest quantity the maximum of the dictionary values (which is the quantity of products) 
        total_stock = sum(product_inventory.values()) #assgining to the total stock the sum of the product quantities(dictionary values)
        print(f"Total number of products: {total_number_of_products}")
        for product in product_inventory: 
            if product_inventory[product] == highest_quantity: #finds the product with the highest quantity
                print(f"{product} has the highest quantity, with a value of {highest_quantity}")
        for product in product_inventory:
            if product_inventory[product] == lowest_quantity: #finds the product with the highest quantity
                print(f"{product} has the lowest quantity with a value of {lowest_quantity}")
        print(f"Total stock available: {total_stock}")
```

```python
#Function that shows the products with low stocks
def low_stock_report():
    low_stock_found = False #setting a variable to be true first
    for product in product_inventory:
        if product_inventory[product] < 5:
            print(f"{product}: {product_inventory[product]}")
            low_stock_found = True #The variable turns true if the product quantity is less than 5
    if low_stock_found == False: #If the product quantity is not less than 5, the variabel remains false and the below statement is printed
        print("All the products have stocks of 5 or greater.")
```

```python
# Menu Function
def inventory_menu():
    try:
        while True:# A loop to repeat the menu as far as one doesn't exit it
            R = int(input("What will you like to do? Click 1 to add product, Click 2 to view inventory, Click 3 to update product quantity, Click 4 to search for a product, Click 5 to remove a product record, Click 6 to check low stock report, Click 7 to show Inventory Statistics, Click 8 to exit"))
            if R ==1:
                add_products()
            elif R ==2:
                view_inventory()
            elif R ==3:
                update_product_record()
            elif R ==4:
                view_product()
            elif R == 5:
                remove_product_record()
            elif R == 6:
                low_stock_report()
            elif R == 7:
                inventory_statistics()
            elif R == 8:
                break
            else: 
                print("Choose between 1-8")
    except ValueError:
        print("Invalid Entry, enter between 1-8")
```
